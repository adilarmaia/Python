```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 100)
y = 2 *x
y2 = x**2
```

```python
plt.plot(x,y2,'o')
plt.show()
```

```python
plt.plot(x,y,'*')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título do gráfico')
plt.show()
```
