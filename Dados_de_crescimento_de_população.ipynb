ESTUDO DE CASOS: CRESCIMENTO DA POPULAÇÃO
-> Crescimentos de 1980 - 2016
-> DataSus

```python
import matplotlib.pyplot as plt

dados = open("/content/drive/MyDrive/populacao_brasileira.csv").readlines()

x = []
y = []

for i in range(len(dados)):
  if i != 0:
    linha = dados[i].split(";")
    x.append(int(linha[0]))
    y.append(int(linha[1]))

plt.bar(x, y, color= "g")
plt.plot(x, y, color='k', linestyle="--")
#plt.scatter(x, y, color="k")

plt.title("Dados de Crescimento populacional Brasileiro de 19980-2016")
plt.xlabel("Ano")
plt.ylabel("População")
#plt.show()
plt.savefig("população.png")
```
